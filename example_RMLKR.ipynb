```python
import sys

from sklearn.manifold import TSNE

import metric_learn
import numpy as np
from sklearn.datasets import make_regression

sys.path.append('/Users/arminsh/Library/CloudStorage/OneDrive-UniversityofIllinois-Urbana/Github_Repo/RMLKR')
import RMLKR

# visualisation imports
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
def plot_tsne(X, y, colormap=plt.cm.Paired):
    plt.figure(figsize=(8, 6))

    # clean the figure
    plt.clf()

    tsne = TSNE(learning_rate='auto', init='random')
    X_embedded = tsne.fit_transform(X)
    plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap=colormap)

    plt.xticks(())
    plt.yticks(())

    plt.show()
```

```python
X_reg, y_reg = make_regression(n_samples=100, n_informative=3, n_features=5,
                               shuffle=True)
```

```python
plot_tsne(X_reg, y_reg, plt.cm.Oranges)
```

```python
rmlkr = RMLKR.RMLKR(lambda_= 5, verbose=True)
X_rmlkr = rmlkr.fit_transform(X_reg, y_reg)
plot_tsne(X_rmlkr, y_reg, plt.cm.Oranges)
```
